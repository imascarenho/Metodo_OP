```python
#Importamos la librería "math" ya que vamos a usar la raíz cuadrada

import math

#Le damos un nombre a la función, (n) va a ser el parametro

def enesimoprimo(n):

# Realizamos una validación por si el numero que se ingresa es igual a 0 ya que devolvería 2 debido a que es el primer numero primo.   
    if n == 0:  #
# Si se ingresa un numero negativo también devuelve el 2
      return "No existe el primo N°0. Ingrese otro número."
    
#Hacemos un docstring que explique que es lo que buscamos con esta función. En nuestro caso encontrar el (n)ésimo número primo.
    
    """
    Calcular el enésimo número primo

    Args:
        n: El número primo que estamos buscando

    Returns:
        El enésimo número primo
    """

#Iniciamos la variable "primo" en 2 ya que es el primer número primo que queremos que tenga en cuenta nuestra función.
#Iniciamos el contador en 1 para poder saber cuantos números primos vamos enconttando
    
    primo = 2
    contador = 1

#Iniciamos un bucle con la función "while", el cual se ejecuta mientras el contador sea menor (<) que (n).
#Hacemos que en cada iteración aumente en 1 el valor de la variable "primo" con el signo (+=). Asumimos que es primo hasta que se demuestre lo contrario.
#"Numprimo" es una bandera con valores booleanos que va a ser decirnos si un número es primo o no, comunica información para asi cambiar la secuencia, empieza a iterar con true pero camnbia al encontrar un divisor.
    
    while contador < n:
        primo += 1
        numprimo = True

#Utilizamos la función "for" para repetir la función un número fijo de veces y "range"para que sea sobre una secuencia de números enteros sucesisvos.
#El "range" empieza en 2 ya que es el primero que queremos tomar.
#La función "math.sqrt" se usa para calcular la raíz cuadrada de "primo", utilizamos "int" para que se tome el número entero.
#Auemntamos en 1 el resultado para asegurarnos de que este incluida la raiz cuadrada
#Probamos divisores hasta la raíz cuadrada de "primo" porque si este tiene un divisor mayor que su raiz cuadrada, debería tambien tener uno menor o igual. 
        
        for i in range(2, int(math.sqrt(primo)) + 1):

#Usamos el módulo (resto) de la division entre "primo" e "i" para comprobar ya que si "primo" es divisible por "i" entonces no es primo. 
#"Break" para romper el bucle cuando "primo" sea divisible por "i" ya que no es primo.
            
            if primo % i == 0:
                numprimo = False
                break  

# Si el número sigue siendo primo (no fue divisible por ningún número), incrementamos el contador en 1, el número de primos encontrados.
        
        if numprimo:
            contador += 1

#Return para que devuelva el enésimo número primo encontrado.        
    return primo  
                
print(enesimoprimo(100000))
```
